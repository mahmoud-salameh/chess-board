```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ChessBoard:
    def __init__(self):
        self.grid = np.ones((8,8,3), np.float64)
        self.red = [0,0]
        self.blue = [0,0]
        div = 1
        for col in range(8):
            for row in range(8):
                div = 0 if div else 1
                if row % 2 == div and col % 2 == div:
                    self.grid[col][row] = np.array((0,0,0))

    def render(self):
        plt.imshow(self.grid)

    def add_red(self,row,col):
        self.grid[row,col] = [1,0,0]
        self.red=[row,col]

    def add_blue(self,row,col):
        self.grid[row,col] = [0,0,1]
        self.blue=[row,col]

    def is_under_attack(self):
        
        if self.blue[0] == self.red[0] or self.blue[1] == self.red[1]:
            return True
        if abs(self.blue[0] - self.red[0]) == abs(self.blue[1] - self.red[1]):
            return True
              
        return False
```

```python
def test_same_row():
    chess = ChessBoard()
    chess.add_red(3,2)
    chess.add_blue(3,7)
    chess.render()
    assert chess.is_under_attack() == True
    print("under attack")

test_same_row()
```

```python
def test_same_column():
    chess = ChessBoard()
    chess.add_red(2,5)
    chess.add_blue(4,5)
    chess.render()
    assert chess.is_under_attack() == True
    print("under attack")

test_same_column()
```

```python
def test_same_diagonal():
    chess = ChessBoard()
    chess.add_red(1,0)
    chess.add_blue(7,6)
    chess.render()
    assert chess.is_under_attack() == True
    print("under attack")

test_same_diagonal()
```

```python
def test_not_under_arrack():
    chess = ChessBoard()
    chess.add_red(6,1)
    chess.add_blue(7,4)
    chess.render()
    assert chess.is_under_attack() == False
    print("Not under attack")

test_not_under_arrack()
```
